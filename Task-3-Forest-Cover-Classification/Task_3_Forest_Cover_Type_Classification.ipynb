# **Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

# **train XGBoost Classifier with default parameters**

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize and train XGBoost Classifier with default parameters
xgb_model = XGBClassifier(objective='multi:softmax', num_class=7, random_state=42)
xgb_model.fit(X_train, y_train - 1)

# Evaluate the model
y_pred_xgb = xgb_model.predict(X_test)
accuracy_xgb = accuracy_score(y_test - 1, y_pred_xgb)
print(f"XGBoost Classifier Accuracy (Default Parameters): {accuracy_xgb}")
print(classification_report(y_test - 1, y_pred_xgb))
```

# **Hyperparameter Tuning for Random Forest classifier**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Random Forest
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5]
}

# Initialize GridSearchCV for Random Forest
grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=3, scoring='accuracy', n_jobs=-1)

# Perform grid search
grid_search_rf.fit(X_train, y_train.values.ravel() - 1)

# Print the best parameters and the best score
print(f"Best parameters for Random Forest: {grid_search_rf.best_params_}")
print(f"Best accuracy for Random Forest: {grid_search_rf.best_score_}")

# Evaluate the best Random Forest model on the test set
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test - 1, y_pred_rf)
print(f"Random Forest Classifier Accuracy (Tuned): {accuracy_rf}")
print(classification_report(y_test - 1, y_pred_rf))
```

# **Hyperparameter Tuning for XGBoost**

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# Define the parameter grid for XGBoost
param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6],
    'learning_rate': [0.1, 0.01]
}

# Initialize GridSearchCV for XGBoost
grid_search_xgb = GridSearchCV(XGBClassifier(objective='multi:softmax', num_class=7, random_state=42), param_grid_xgb, cv=3, scoring='accuracy', n_jobs=-1)

# Perform grid search
grid_search_xgb.fit(X_train, y_train.values.ravel() - 1)

# Print the best parameters and the best score
print(f"Best parameters for XGBoost: {grid_search_xgb.best_params_}")
print(f"Best accuracy for XGBoost: {grid_search_xgb.best_score_}")

# Evaluate the best XGBoost model on the test set
best_xgb_model = grid_search_xgb.best_estimator_
y_pred_xgb_tuned = best_xgb_model.predict(X_test)
accuracy_xgb_tuned = accuracy_score(y_test - 1, y_pred_xgb_tuned)
print(f"XGBoost Classifier Accuracy (Tuned): {accuracy_xgb_tuned}")
print(classification_report(y_test - 1, y_pred_xgb_tuned))
```

# **Random Forest vs. XGBoost**

```python
# Evaluate the best Random Forest model on the test set
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test - 1, y_pred_rf)
print(f"Random Forest Classifier Accuracy (Tuned): {accuracy_rf}")
print(classification_report(y_test - 1, y_pred_rf))

# Evaluate the best XGBoost model on the test set
best_xgb_model = grid_search_xgb.best_estimator_
y_pred_xgb_tuned = best_xgb_model.predict(X_test)
accuracy_xgb_tuned = accuracy_score(y_test - 1, y_pred_xgb_tuned)
print(f"XGBoost Classifier Accuracy (Tuned): {accuracy_xgb_tuned}")
print(classification_report(y_test - 1, y_pred_xgb_tuned))
```

# **Visualize confusion matrix and feature importance**

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Confusion Matrix for Tuned Random Forest
cm_rf = confusion_matrix(y_test - 1, y_pred_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (Tuned Random Forest)')
plt.xlabel('Predicted Cover Type')
plt.ylabel('Actual Cover Type')
plt.show()

# Feature Importance for Tuned Random Forest
feature_importance_rf = pd.Series(best_rf_model.feature_importances_, index=X_train.columns).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
feature_importance_rf.plot(kind='bar')
plt.title('Feature Importance (Tuned Random Forest)')
plt.ylabel('Importance')
plt.tight_layout()
plt.show()

# Confusion Matrix for Tuned XGBoost
cm_xgb = confusion_matrix(y_test - 1, y_pred_xgb_tuned)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (Tuned XGBoost)')
plt.xlabel('Predicted Cover Type')
plt.ylabel('Actual Cover Type')
plt.show()

# Feature Importance for Tuned XGBoost
feature_importance_xgb = pd.Series(best_xgb_model.feature_importances_, index=X_train.columns).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
feature_importance_xgb.plot(kind='bar')
plt.title('Feature Importance (Tuned XGBoost)')
plt.ylabel('Importance')
plt.tight_layout()
plt.show()
```
